```python
#import library
import pandas as pd
#read information into Dataframe
airbnb_data = pd.read_csv('airbnb_data.csv')
booking_data = pd.read_excel('ScrapingBooking.xlsx')
print(airbnb_data.info())
print(booking_data.info())
```

Bearbeitung von AirBnB Datensatz

```python
#Vorbereitung AirBnB Datensatz
airbnb_clean = airbnb_data.drop(['end_buch', 'title'], axis=1)
airbnb_clean.info()
```

```python
airbnb_clean = airbnb_clean.iloc[:, [0,5,1,6,2,7,3,4]]
airbnb_clean.info()
```

```python
airbnb_clean.head()
```

```python
airbnb_clean = airbnb_clean.rename(index=str, columns={'date_crawl':'date_crawl' , 'ort':'city', 'link':'uid', 'start_buch':'book_date', 'price':'price', 'event':'event', 'rating':'rating', 'provider':'provider'})
airbnb_clean.info()
```

```python
rating = airbnb_clean['rating'].str.split(' ', expand=True)
airbnb_clean['rating'] = rating[0]
airbnb_clean.head()
```

```python
price = airbnb_clean['price'].str.split('\xa0', expand=True)
airbnb_clean['price'] = price[0]
airbnb_clean.head()
```

```python
airbnb_clean['provider'] = airbnb_clean['provider'].str.replace('Private:r Gastgeber:in', 'privat')
airbnb_clean['provider'] = airbnb_clean['provider'].str.replace('Gewerbliche:r Vermieter:in', 'gewerblich')
airbnb_clean['event'] = airbnb_clean['event'].str.replace('ja', 'True')
airbnb_clean['event'] = airbnb_clean['event'].str.replace('nein', 'False')
airbnb_clean['rating'] = airbnb_clean['rating'].str.replace('NaN', '')
airbnb_clean['price'] = airbnb_clean['price'].str.replace('NaN', '')
airbnb_clean['rating'] = airbnb_clean['rating'].str.replace('Neu', '')
airbnb_clean['price'] = airbnb_clean['price'].str.replace("’", '')
airbnb_clean
```

```python
airbnb_clean.isnull().sum()
```

```python
airbnb_clean_nonull = airbnb_clean.drop(airbnb_clean[airbnb_clean['city'].isnull() == True].index)
airbnb_clean_nonull
```

```python
airbnb_clean_nonull.isnull().sum()
```

```python
airbnb_clean_nonull = airbnb_clean_nonull.drop(airbnb_clean_nonull[airbnb_clean_nonull['provider'].isnull() == True].index)
airbnb_clean_nonull.isnull().sum()
```

```python
airbnb_clean_nonull['date_crawl'] = airbnb_clean_nonull['date_crawl'].astype('datetime64[ns]')
airbnb_clean_nonull['book_date'] = airbnb_clean_nonull['book_date'].astype('datetime64[ns]')
airbnb_clean_nonull['price'] = pd.to_numeric(airbnb_clean_nonull['price'])
airbnb_clean_nonull['rating'] = pd.to_numeric(airbnb_clean_nonull['rating'])
airbnb_clean_nonull['event'] = airbnb_clean_nonull['event'].astype('bool')
airbnb_clean_nonull.info()
```

```python
airbnb_clean_copy = airbnb_clean_nonull
airbnb_clean_copy['rating'] = airbnb_clean_copy['rating'].fillna(airbnb_clean_copy['rating'].mean())
airbnb_clean_copy['price'] = airbnb_clean_copy['price'].fillna(airbnb_clean_copy['price'].mean())
airbnb_clean_nonull['price'] = airbnb_clean_nonull['price'].astype('int')
airbnb_clean_copy.info()
```

Bearbeitung von Booking Datensatz

```python
print(booking_data.info())
```

```python
booking_clean = booking_data.drop('Name', axis=1)
booking_clean.info()
```

```python
booking_clean = booking_clean.iloc[:, [1,0,7,4,2,3,5,6]]
booking_clean.info()
```

```python
booking_clean = booking_clean.rename(index=str, columns={'Datum scraped':'date_crawl' , 'Stadt':'city', 'Link':'uid', 'Datum':'book_date', 'Preis':'price', 'Event Ja/Nein':'event', 'Bewertung':'rating', 'Anbieter':'provider'})
booking_clean.info()
```

```python
booking_clean['rating'] = booking_clean['rating'].str.replace(',', '.')
booking_clean['rating'] = pd.to_numeric(booking_clean['rating'])
booking_clean
```

```python
booking_clean['rating'] = booking_clean['rating']/2
booking_clean['rating'] = booking_clean['rating'].round(2)
booking_clean
```

```python
booking_clean.isnull().sum()
```

```python
booking_clean.info()
```

```python
booking_clean['date_crawl'] = booking_clean['date_crawl'].astype('datetime64[ns]')
booking_clean['book_date'] = booking_clean['book_date'].astype('datetime64[ns]')
booking_clean['event'] = booking_clean['event'].str.replace('ja', 'True')
booking_clean['event'] = booking_clean['event'].str.replace('nein', 'False')
booking_clean['event'] = booking_clean['event'].astype('bool')
booking_clean.info()
```

```python
booking_clean_copy = booking_clean
booking_clean_copy['rating'] = booking_clean_copy['rating'].fillna(booking_clean_copy['rating'].mean())
booking_clean_copy.info()
```

Merge Data into one dataset

```python
dataframes = [airbnb_clean_copy,booking_clean_copy]
dataset = pd.concat(dataframes)
dataset['city'] = dataset['city'].str.replace('Br√ºssel', 'Bruessel')
dataset['city'] = dataset['city'].str.replace('M√ºnchen', 'Muenchen')
dataset['city'] = dataset['city'].str.replace('Monaco-City', 'Monaco')
#drop unrealistische Preise
dataset.drop(dataset[dataset['price'] <= 20].index, inplace = True)
dataset.info()
```

```python
dataset.to_excel('merged_dataset.xlsx')
```
