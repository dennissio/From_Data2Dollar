```python
import pandas as pd
```

```python
# Excel-Dateien einlesen und DF erstellen
merged_dataset = pd.read_excel('merged_dataset.xlsx')
tweets_cleaned = pd.read_excel('tweets_cleaned.xlsx')
```

```python
#Blick in die Hotel-Daten
merged_dataset
```

```python
#Blick in die Twitter-Daten
tweets_cleaned
```

```python
# leere Spalten für die Twitter-Elemente zum 'merged_dataset' DF hinzufügen
merged_dataset['event_name'] = None
merged_dataset['tweets'] = None
merged_dataset['retweets'] = None
merged_dataset['likes'] = None
merged_dataset['replies'] = None
```

```python
#Blick in die Daten
merged_dataset
```

```python
# Die DataFrames basierend auf der 'city'-Spalte kombinieren, sodass in den neu hinzugefügten Spalten, die gleichen Werte ffür jede Stadt stehen
# Dies wird für alle Städte wiederholt
combined_df = merged_dataset.copy()
for index, row in tweets_cleaned.iterrows():
    city = row['city']
    mask = combined_df['city'] == city
    combined_df.loc[mask, 'event_name'] = row['event_name']
    combined_df.loc[mask, 'tweets'] = row['tweets']
    combined_df.loc[mask, 'retweets'] = row['retweets']
    combined_df.loc[mask, 'likes'] = row['likes']
    combined_df.loc[mask, 'replies'] = row['replies']
```

```python
# Blick in den neuen DF
combined_df
```

```python
# DF in einer neuen CSV-Datei speichern
combined_df.to_csv('data_combined.csv', index=False, encoding='utf-8-sig', sep=',')
```
