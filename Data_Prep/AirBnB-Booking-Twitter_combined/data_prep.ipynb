```python
import pandas as pd
```

```python
# Excel-Dateien einlesen und DF erstellen
merged_dataset = pd.read_excel('merged_dataset.xlsx')
tweets_cleaned = pd.read_excel('tweets_cleaned.xlsx')
tweets_cleaned_aggr = pd.read_csv('tweets_cleaned_aggr.csv')
```

```python
#Blick in die Twitter-Daten
tweets_cleaned
```

```python
#Blick in die aggregierten Twitter-Daten
tweets_cleaned_aggr
```

```python
# leere Spalten fÃ¼r die Twitter-Elemente zum 'merged_dataset' DF hinzufÃ¼gen
merged_dataset['event_name'] = None
merged_dataset['tweets'] = None
merged_dataset['retweets'] = None
merged_dataset['likes'] = None
merged_dataset['replies'] = None
```

```python
#Blick in die Daten
merged_dataset
```

```python
# Die DataFrames basierend auf der 'city'-Spalte kombinieren, sodass in den neu hinzugefÃ¼gten Spalten, die gleichen Werte fÃ¼r jede Stadt stehen
# Dies wird fÃ¼r alle StÃ¤dte wiederholt
combined_df = merged_dataset.copy()
for index, row in tweets_cleaned.iterrows():
    city = row['city']
    mask = combined_df['city'] == city
    combined_df.loc[mask, 'event_name'] = row['event_name']
    combined_df.loc[mask, 'tweets'] = row['tweets']
    combined_df.loc[mask, 'retweets'] = row['retweets']
    combined_df.loc[mask, 'likes'] = row['likes']
    combined_df.loc[mask, 'replies'] = row['replies']
```

```python
#Blick in die Daten
combined_df
```

```python
# 4 neue Spalten mit Social Media Analytics Kennzahlen und der jewewiligen Berechnung erstellen
# Es wird nebenbei sichergestellt, dass kein Teilen durch Null oder None auftritt.
combined_df['amplification_rate'] = combined_df.apply(lambda row: row['retweets'] / row['tweets'] if row['tweets'] != 0 and row['tweets'] is not None else 0, axis=1)
combined_df['conversation_rate'] = combined_df.apply(lambda row: row['replies'] / row['tweets'] if row['tweets'] != 0 and row['tweets'] is not None else 0, axis=1)
combined_df['applause_rate'] = combined_df.apply(lambda row: row['likes'] / row['tweets'] if row['tweets'] != 0 and row['tweets'] is not None else 0, axis=1)
combined_df['total_engagement'] = combined_df['likes'] + combined_df['tweets'] + combined_df['replies'] + combined_df['retweets']
```

```python
#Blick in die Daten
combined_df
```

```python
# Basierend auf Berechnungen der Quantile in der Excel-Datei zu den Events, haben wir 4 Kategorien erstellt und die StÃ¤dte entsprechend ihrer "Relevanz" (oder Einfluss) eingeteil
# Eine Funktion get_relevance wird erstellt, welche diese Einteilung vornimmt. Im Anschluss wird diese auf das DF angewandt
def get_relevance(city):
    if city in ["Barcelona", "Muenchen", "La Palma"]:
        return "gering"
    elif city in ["Lyon", "Halle", "Bologna", "Helsinki", "Bruessel", "Sevilla"]:
        return "mittel"
    elif city in ["Berlin", "Brighton", "Hamburg"]:
        return "hoch"
    elif city in ["Amsterdam", "Monaco", "Noordwijk"]:
        return "sehr hoch"
    else:
        return "nicht definiert"

combined_df["relevance"] = combined_df["city"].apply(get_relevance)
```

```python
#Blick in die Daten
combined_df
```

```python
# Neue Spalten in combined_df erstellen und mit None initialisieren
combined_df[['retweets_cum', 'likes_cum', 'replies_cum', 'retweets_cont', 'likes_cont', 'replies_cont']] = None

# Die Dataframes "tweets_cleaned_aggr" und "combined_df" kombinieren
for index, row in tweets_cleaned_aggr.iterrows():
    date_condition = combined_df['date_crawl'] == row['date']
    hashtag_condition = combined_df['event_name'] == row['hashtag']
    
    combined_df.loc[date_condition & hashtag_condition, ['retweets_cum', 'likes_cum', 'replies_cum', 'retweets_cont', 'likes_cont', 'replies_cont']] = row[['retweets_cum', 'likes_cum', 'replies_cum', 'retweets', 'likes', 'replies']].values
```

```python
# Blick in den neuen DF
combined_df
```

```python
# DF in einer neuen CSV-Datei speichern
combined_df.to_csv('data_combined.csv', index=False, encoding='utf-8-sig', sep=',')
```
